```python
import pandas as pd
```

```python
df=pd.read_csv('income.csv')
```

```python
df
```

```python
import matplotlib.pyplot as plt
```

```python
plt.xlabel('Age')
plt.ylabel('Income($)')
plt.scatter(df['Age'],df['Income($)'],color='red',marker='x')
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
```

```python
km=KMeans(n_clusters=3)
```

```python
scaler=MinMaxScaler()
```

```python
scaler.fit(df[['Income($)']])
```

```python
df['Income($)']=scaler.transform(df[['Income($)']])
```

```python
df
```

```python
plt.xlabel('Age')
plt.ylabel('Income($)')
plt.scatter(df['Age'],df['Income($)'],color='red')
```

```python
predicted=km.fit_predict(df[['Age','Income($)']])
```

```python
predicted
```

```python
df['Cluster']=predicted
```

```python
df
```

```python
df0=df[df.Cluster==0]
df1=df[df.Cluster==1]
df2=df[df.Cluster==2]
```

```python
plt.xlabel('Age')
plt.ylabel('Income($)')
plt.scatter(df0.Age,df0['Income($)'],color='green')
plt.scatter(df1.Age,df1['Income($)'],color='blue')
plt.scatter(df2.Age,df2['Income($)'],color='red')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='purple',marker='x')
plt.show()
```
